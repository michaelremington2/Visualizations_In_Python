```python
%matplotlib notebook
# Plot Grid
plt.style.use('fivethirtyeight')

x = list(np.arange(-3,3,0.1))
multi=[1,2,3,4]

fig, axs = plt.subplots(len(multi), 1, sharex=True, sharey=False)

for i, value in enumerate(multi):
    y1 = [math.sin(value*i_) for i_ in x]
    y2 = [math.cos(value*i_) for i_ in x]
    axs[i].plot(x, y1)
    axs[i].set_title(f'sin({value}x) vs cos({value}x)')
    axs[i].plot(x, y2)
    axs[i].set_ylim([-1.2,1.2])
    
plt.tight_layout()
plt.show()
```

```python
%matplotlib notebook
#######
# Other Types of Figures: Bar Charts
#######

students = ['Mary', 'Dave', 'Michael', 'Sara']
grades = list(np.random.uniform(60,100,4))
chart = plt.bar(students, grades)

for bar in chart:
   height = bar.get_height()
   plt.annotate('{}'.format(round(height)),
      xy=(bar.get_x() + bar.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()
```

```python
%matplotlib notebook
exam = ("Exam 1", "Exam 2", "Exam 3")
student_scores = {
    'Mary': (90,81,85),
    'Dave': (60, 75, 72),
    'Michael': (91, 95, 99),
    'Sara': (58, 66, 60)
    
}

x = np.arange(len(exam))  # the label locations
width = 0.15  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(constrained_layout=True)

for student, exam_scores in student_scores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, exam_scores, width, label=student)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Grade')
ax.set_title('Exam Grades for the Semester')
ax.set_xticks(x + width, exam)
ax.legend(loc='lower center')
ax.set_ylim(0, 105)

plt.show()
```

```python
###################
### Multi demensional data: Heat maps
###################
#plt.style.use('seaborn')
# Figure from https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```

```python
########
# Making a function for repeatbility
#######

def make_heatmap(x_labels,y_labels,data,title=None):
    fig, ax = plt.subplots()
    im = ax.imshow(data)

    # Show all ticks and label them with the respective list entries
    ax.set_xticks(np.arange(len(x_labels)), labels=x_labels)
    ax.set_yticks(np.arange(len(y_labels)), labels=y_labels)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    for i in range(len(x_labels)):
        for j in range(len(y_labels)):
            text = ax.text(j, i, data[i, j],
                           ha="center", va="center", color="w")
    if title is not None:
        ax.set_title(title)
    fig.tight_layout()
    plt.show()
```

```python
#####
# using our function
#####
make_heatmap(farmers, vegetables, data=np.transpose(harvest), title = 'Crop Harvest Amount By Farmer')
```

```python
%matplotlib notebook
#######
## Multidemensional data: 3D Figures
#######

#https://www.python-graph-gallery.com/372-3d-pca-result
# https://stackoverflow.com/questions/34748798/python-legend-in-3dplot

from mpl_toolkits.mplot3d import Axes3D

x = [math.sin(i) for i in np.arange(-10,10,0.1)]
y = [math.cos(i) for i in np.arange(-10,10,0.1)]
z = [math.cos(i)+math.sin(i) for i in np.arange(-10,10,0.1)]
z2 = [math.cos(2*i)+math.sin(2*i) for i in np.arange(-10,10,0.1)]


fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')



ax.scatter(x, y, z,label='cos(x)+sin(x)')
ax.scatter(x, y, z2,label='cos(2x)+sin(2x)')

ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("cos(x)+sin(x)")

plt.show()

```

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
# Start with a square Figure.
x = np.random.randn(1000)
y = np.random.randn(1000)
fig = plt.figure(figsize=(6, 6))
# Add a gridspec with two rows and two columns and a ratio of 1 to 4 between
# the size of the marginal axes and the main axes in both directions.
# Also adjust the subplot parameters for a square plot.
gs = fig.add_gridspec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                      left=0.1, right=0.8, bottom=0.1, top=0.9,
                      wspace=0.05, hspace=0.05)
# Create the Axes.
ax = fig.add_subplot(gs[1, 0])
ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)
# Draw the scatter plot and marginals.
scatter_hist(x, y, ax, ax_histx, ax_histy)
```
