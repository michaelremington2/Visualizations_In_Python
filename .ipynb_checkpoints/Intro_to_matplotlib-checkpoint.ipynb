```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random as rand
x = list(np.arange(-3,3,0.1))
y1 = [math.sin(i) for i in x]
y2 = [math.cos(i) for i in x]
```

```python
plt.plot(x,y1)
plt.show()
```

```python
%matplotlib notebook
plt.plot(x,y1)
plt.show()
```

```python
%matplotlib notebook
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
%matplotlib notebook
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Cos(x) vs Sin(x)')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
%matplotlib notebook

plt.plot(x, y1, color='r', marker = '.', linewidth=3, label='sin(x)')
plt.plot(x, y2, color='#89CFF0', marker = 'o', label='cos(x)')
plt.title('Cos(x) vs Sin(x)')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
#plt.legend(loc='upper right')
#plt.grid()
plt.tight_layout()
plt.show()
```

```python
print(plt.style.available)
```

```python
%matplotlib notebook
plt.style.use('seaborn-colorblind')
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Cos(x) vs Sin(x)')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
# Plot Grid
fig, axs = plt.subplots(2, 1, figsize=(8, 4))


axs[0].plot(x, y1)
axs[0].set_title('sin(x)')

axs[1].plot(x, y2)
axs[1].set_title('cos(x)')
axs[1].set_ylim([-1.5,1.5])
plt.tight_layout()
```

```python
# Plot Grid
plt.style.use('fivethirtyeight')

x = list(np.arange(-3,3,0.1))
multi=[1,2,3,4]

fig, axs = plt.subplots(len(multi), 1, sharex=True, sharey=False)

for i, value in enumerate(multi):
    y1 = [math.sin(value*i) for i in x]
    y2 = [math.cos(value*i) for i in x]
    axs[i].plot(x, y1)
    axs[i].set_title(f'sin({value}x) vs cos({value}x)')
    axs[i].plot(x, y2)
    axs[i].set_ylim([-1.2,1.2])
    
plt.tight_layout()
plt.show()
```

```python
#######
# Other types of Figures: Scatter plot
#######
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.4)
plt.show()
```

```python
%matplotlib notebook
#######
# Other Types of Figures: Bar Charts
#######

students = ['Mary', 'Dave', 'Michael', 'Sara']
grades = list(np.random.uniform(60,100,4))
chart = plt.bar(students, grades)

for bar in chart:
   height = bar.get_height()
   plt.annotate('{}'.format(round(height)),
      xy=(bar.get_x() + bar.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()
```

```python

exam = ("Exam 1", "Exam 2", "Exam 3")
student_scores = {
    'Mary': (90,81,85),
    'Dave': (60, 75, 72),
    'Michael': (91, 95, 99),
    'Sara': (58, 66, 60)
    
}

x = np.arange(len(exam))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(constrained_layout=True)

for attribute, measurement in student_scores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Grade')
ax.set_title('Exam Grades for the Semester')
ax.set_xticks(x + width, exam)
ax.legend(loc='lower center')
ax.set_ylim(0, 105)

plt.show()
```
