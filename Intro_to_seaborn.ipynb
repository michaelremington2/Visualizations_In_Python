# Seaborn

Seaborn is a graphing package that sits on top of matplotlib that allows the user to easily explore data. Data inputs include pandas dataframes or numpy arrays. It is useful for making beautiful plots for data analysis quickly and intuitively compared to matplotlib. You will still need to know a lot of principles of matplotlib but seaborn simplifies making complex, customized figures into much simplier steps.

Lets dive in!

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random as rand
import pandas as pd
```

```python
x = list(np.arange(-3,3,0.1))
y1 = [math.sin(i) for i in x]
y2 = [math.cos(i) for i in x]
d = {"x": x,
     "sin": y1,
     "cos": y2}
df = pd.DataFrame(d)
df.head()
```

```python
sns.lineplot(data=df, x="x", y="sin")
```

```python
fig, axs = plt.subplots(1, 2, sharex=False,sharey=True, figsize=(8,4))

f1 = sns.lineplot(data=df, ax=axs[0], x="x", y="sin").set(
                                                  title = "sin(x)",
                                                  ylabel = "y",
                                                  xlabel = "x",
                                                  ylim=(-1.1, 1.1))
f2 = sns.lineplot(data=df, ax=axs[1], x="x", y="cos").set(
                                                  title = "cos(x)",
                                                  ylabel = "y",
                                                  xlabel = "x",
                                                  ylim=(-1.1, 1.1))
axs[1].grid()
```

```python
#####
## Load Iris Data
#####
iris = pd.read_csv("Data/Iris.csv")
iris.drop('Id', inplace=True, axis=1)
iris.head()
```

```python
%matplotlib notebook
sns.set_style("darkgrid")
sns.scatterplot(data=iris, x='SepalWidthCm', y='SepalLengthCm', hue="Species")
```

```python
%matplotlib notebook
sns.set(rc={'figure.figsize':(8,4)})
graph = sns.boxplot(x="Species", y='SepalLengthCm', data=iris)
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
cols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for variable in cols:
    plt.figure(figsize=(8,4))
    ax = sns.violinplot(x="Species", y=variable, data=iris)
    plt.title(variable, fontdict=font)
    plt.xlabel('Species', fontdict=font)
    plt.ylabel(variable, fontdict=font)
    
```

```python
%matplotlib notebook
# calculate the correlation matrix
corr = iris.corr()

# plot the heatmap
plt.figure(figsize=(8,8))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

```python
%matplotlib notebook
######
## Histograms again
######
sns.histplot(data=iris, x='SepalLengthCm')
```

```python
%matplotlib notebook
sns.histplot(data=iris, x='SepalLengthCm', binwidth=1)
```

```python
%matplotlib notebook
sns.histplot(data=iris, x='SepalLengthCm', bins=30)
```

```python
gridkw = dict(height_ratios=[1, 4], width_ratios=[4,1])
fig, (ax1, ax2) = plt.subplots(2, 2, gridspec_kw=gridkw)

# the scatter plot:
sns.scatterplot(data=iris, ax=ax2[0], x='SepalWidthCm', y='SepalLengthCm')
sns.histplot(data=iris, ax=ax1[0], x='SepalWidthCm').set(xlabel="")
sns.histplot(data=iris, ax=ax2[1], y='SepalLengthCm').set(ylabel="")
ax1[1].axis('off')
plt.show()
```
